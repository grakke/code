### 数据结构

* 列表 list 列表
    - 动态:需要存储指针指向对应的元素（对于 int 型，8 字节）
    - 可变:需要额外存储已经分配的长度大小（8 字节）
    - 为了减小每次增加/删减操作时空间分配的开销，每次分配空间时都会额外多分配一些，这样的机制（over-allocating）保证了其操作的高效性：增加 / 删除的时间复杂度均为 O(1)
    * 动态（mutable）长度不固定，可以增加删除或修改
    * 初始化
        - list() 一个 function call，会创建stack，并且进行一系列参数检查的操作
        - [] 内置C函数，可以直接被调用
* 元组
    - 长度大小固定，元素不可变，所以存储空间固定
    - 对于一些静态变量，比如元组，如果不被使用并且占用空间不大时，Python 会暂时缓存这部分内存。下次再创建同样大小的元组时，就可以不用再向操作系统发出请求，去寻找内存，而是可以直接分配之前缓存的内存空间，大大加快程序的运行速度
* 内部实现 都是array形式
    - list 可变 一个over-allocate的array
    - tuple 不可变 长度大小固定

* 索引操作:两者的速度差别非常小


````python
# 在每次调用next()的时候执行，遇到yield语句返回，再次执行时从上次返回的yield语句处继续执行
def fib(max):
    n, a, b = 0, 0, 1
    while n < max:
        yield b
        a, b = b, a + b
        n = n + 1
    return 'done'


g = fib(5)
while True:
    try:
        x = next(g)
        print('g:', x)
    except StopIteration as e:
        print('Generator return value:', e.value)
        break
````

````python
def triangles():
    l = [1]

    while True:
        yield l
        l = [1] + [l[i] + l[i + 1] for i in range(len(l) - 1) if len(l) > 1] + [1]
l
````
