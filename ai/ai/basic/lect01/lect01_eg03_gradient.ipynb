# 梯度下降算法

```python
import numpy as np
```

```python
def func(x):
    return (x + 5) ** 2

x_points = np.linspace(-10, 5, num=100)
y_points = func(x_points)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.scatter(x_points, y_points)
plt.show()
```

```python
import random

start_x = random.uniform(-20, 10)  # 起始点
alpha = 0.01  # 学习率
max_iter = 1000  # 最大迭代次数

# 梯度：表示变化大小与方向
def gradient(x):
    return 2 * (x + 5)
```

```python
n_iter = 0
cur_x = start_x
x_list = []
y_list = []
while n_iter < max_iter:
    prev_x = cur_x
    # 按照梯度的变化加权
    cur_x = prev_x - alpha * gradient(prev_x)
    n_iter += 1
    
    cur_y = func(cur_x)
    x_list.append(cur_x)
    y_list.append(cur_y)
    print('迭代次数={}, x={}, y={}'.format(n_iter, cur_x, cur_y))
```

```python
# 可视化优化过程
plt.figure()
plt.scatter(range(max_iter), y_list)
plt.show()
```
